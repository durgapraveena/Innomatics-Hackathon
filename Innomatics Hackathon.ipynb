```python
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'ratings' with columns 'movieId', 'title', and 'rating'
# Load the data from your CSV file or any other source
ratings = pd.read_csv('merged.csv')

# Filter ratings for "Fight Club (1999)"
fight_club_ratings = ratings[ratings['title'] == 'Fight Club (1999)']

# Plot a histogram of user ratings
plt.figure(figsize=(10, 6))
plt.hist(fight_club_ratings['rating'], bins=5, edgecolor='black', alpha=0.7)
plt.title('Distribution of User Ratings for Fight Club (1999)')
plt.xlabel('Rating')
plt.ylabel('Number of Ratings')
plt.show()
```

```python
import pandas as pd

# Assuming you have a DataFrame named 'ratings' with columns 'movieId', 'title', and 'rating'
# Load the data from your CSV file or any other source
ratings = pd.read_csv('merged.csv')

# Calculate the average rating for each movie
average_ratings = ratings.groupby('title')['rating'].mean()

# Find the movie with the highest average rating
most_popular_movie = average_ratings.idxmax()
highest_average_rating = average_ratings.max()

print("Most popular movie based on average user ratings:")
print("Movie Title:", most_popular_movie)
print("Average Rating:", highest_average_rating)
```

```python
import pandas as pd

# Assuming you have a DataFrame named 'ratings' with columns 'movieId', 'title', and 'rating'
# Load the data from your CSV file or any other source
ratings = pd.read_csv('merged.csv')

# Count the number of ratings for each movie
ratings_count = ratings.groupby('title')['rating'].count()

# Sort the movies based on the number of ratings in descending order
top_movies = ratings_count.sort_values(ascending=False).head(5)

# Display the top 5 movies
print("Top 5 popular movies based on number of user ratings:")
print(top_movies)
```

```python
import pandas as pd

# Assuming you have a DataFrame named 'ratings' with columns 'movieId', 'title', and 'rating'
# and a DataFrame named 'movies' with information about genres
# Load the data from your CSV file or any other source
ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')  # Assuming you have a 'genres' column in this DataFrame

# Merge the ratings and movies DataFrames
merged_data = pd.merge(ratings, movies, on='movieId')

# Filter for Sci-Fi movies
sci_fi_movies = merged_data[merged_data['genres'].str.contains('Sci-Fi')]

# Count the number of ratings for each Sci-Fi movie
ratings_count_sci_fi = sci_fi_movies.groupby('title')['rating'].count()

# Sort the Sci-Fi movies based on the number of ratings in descending order
sorted_sci_fi_movies = ratings_count_sci_fi.sort_values(ascending=False)

# Identify the third most popular Sci-Fi movie
third_most_popular_sci_fi_movie = sorted_sci_fi_movies.index[2]

print("Third most popular Sci-Fi movie based on number of user ratings:")
print(third_most_popular_sci_fi_movie)
```
